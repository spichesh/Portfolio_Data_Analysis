* Campaign_id - ID кампании;
* Geo - геопозиция, на которую настроена рекламная кампания;
* Buyer - байер, выполнивший запуск;
* Impressions - количество показов рекламных объявлений;
* Clicks - количество кликов; 
* Installs - количество установок;
* Registrations - количество регистраций;
* Spend - Затраченная сумма;
* Revenue_ad (1d/7d/30d) - Выручка с рекламы от пользователей, зарегистрировавшихся в приложении, на 1/7/30 день их жизни
* Revenue_in_app (1d/7d/30d) - Выручка с покупок от пользователей, зарегистрировавшихся в приложении, на 1/7/30 день их жизни
* Payers (1d/7d/30d) - Количество покупателей в приложении на 1/7/30 день их жизни

Главные вопросы рекламной кампании:
* Какова стоимость привлечения покупателя?
* Какое отношение выручки с покупателей к затратам на рекламную кампанию? 

```python
#посчитаю основные метрики

data['CR_Installs'] = data.Installs/data.Impressions * 100 #конверсия из показов в установку приложения

data['CR_Registrations'] = data.Registrations/data.Installs * 100 #конверсия в зарегистрированного пользователя

data['CTR'] = data.Clicks/data.Impressions * 100 #показатель кликабельности

data['CPI'] = data.Spend/data.Clicks #цена за клик

data['CP_Installs'] = data.Spend/data.CR_Installs #стоимость конверсии в установку

data['CP_Registrations'] = data.Spend/data.CR_Registrations #стоимость конверсии в регистрацию

data['CAC1d'] = data.Spend/data.Payers1d #стоимость привлечения покупателя
data['CAC7d'] = data.Spend/data.Payers7d #стоимость привлечения покупателя
data['CAC30d'] = data.Spend/data.Payers30d #стоимость привлечения покупателя

data['RTS30d'] = (data.Revenue_ad30d + data.Revenue_in_app30d)/data.Spend  #Revenue Through Spend - отношение выручки с покупателей к затратам
data['RTS7d'] = (data.Revenue_ad7d + data.Revenue_in_app7d)/data.Spend  #Revenue Through Spend - отношение выручки с покупателей к затратам
data['RTS1d'] = (data.Revenue_ad1d + data.Revenue_in_app1d)/data.Spend  #Revenue Through Spend - отношение выручки с покупателей к затратам
data['Revenue_30d'] = data.Revenue_ad30d + data.Revenue_in_app30d
data['Revenue_7d'] = data.Revenue_ad7d + data.Revenue_in_app7d
data['Revenue_1d'] = data.Revenue_ad1d + data.Revenue_in_app1d
```

```python
data.head()
```

```python
data.replace(np.inf, 0, inplace = True)
```

## Оценка окупаемости рекламной кампании

#### 1. В разрезе по медиабайерам

```python
buyers_revenue = data.groupby('Buyer', as_index = False)[['Revenue_30d', 'Revenue_7d', 'Revenue_1d', 'Spend']].sum()
```

```python
buyers_revenue['RTS30d'] = buyers_revenue.Revenue_30d/buyers_revenue.Spend
buyers_revenue['RTS7d'] = buyers_revenue.Revenue_7d/buyers_revenue.Spend
buyers_revenue['RTS1d'] = buyers_revenue.Revenue_7d/buyers_revenue.Spend
```

```python
buyers_revenue[['Buyer', 'RTS30d', 'RTS7d', 'RTS1d']].sort_values(by = 'RTS30d', ascending = False)
```

По результатам рекламных кампаний только buyer_3 окупился.
Наименее результативной была кампания buyer_5

```python
data.groupby('Buyer')['Spend'].sum().sort_values()
```

```python
data[data.Buyer == 'buyer_5'].groupby('Geo', as_index = False)['CAC30d', 'Spend'].mean().sort_values(by ='Spend')
```

buyer_5 не смог привлечь платежеспособных клиентов из большинства гео. Возможно, люди в этих местах в целом не интересуются продуктом.

#### 2. В разрезе по гео

```python
geo_revenue = data.groupby('Geo', as_index = False)[['Revenue_30d', 'Revenue_7d', 'Revenue_1d', 'Spend']].sum()
```

```python
geo_revenue['RTS30d'] = geo_revenue.Revenue_30d/geo_revenue.Spend
geo_revenue['RTS7d'] = geo_revenue.Revenue_7d/geo_revenue.Spend
geo_revenue['RTS1d'] = geo_revenue.Revenue_7d/geo_revenue.Spend
geo_revenue[['Geo', 'RTS30d', 'RTS7d', 'RTS1d']].sort_values(by = 'RTS30d', ascending = False)
```

Самая высокая окупаемость у Западной Европы и Северной Америки.

 Самая низкая у Австралии и Океании, Северной Африки. Необходимо пересмотреть проект рекламной кампании для этох гео.

Также, очень большая стоимость пользователей из стран Западной Европы.
Самые 'дешевые' пользователи находятся в Латинской Америке.

### Анализ конверсии в установку и регистрацию

```python
results_campains = data.groupby('Geo', as_index = False).agg({
    'CPI': 'mean',
    'CR_Installs': 'mean',
    'CR_Registrations': 'mean',
    'CP_Installs': 'mean',
    'CP_Registrations': 'mean'   
})
results_campains
```

```python
fig, ax = plt.subplots(figsize =(10, 8))

ax.barh(results_campains.sort_values(by = 'CR_Installs').Geo, results_campains.sort_values(by = 'CR_Installs').CR_Installs)

for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
ax.invert_yaxis()
 
for i in ax.patches:
    plt.text(i.get_width()+0.01, i.get_y()+0.5, 
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
ax.set_title('Конверсия в установку по гео',
             loc ='left');
```

* Конверсия в установку хуже всего была у кампании с гео "australia oceania"
* Лучшая - "north america"

```python
fig, ax = plt.subplots(figsize =(10, 8))

ax.barh(results_campains.sort_values(by = 'CR_Registrations').Geo, results_campains.sort_values(by = 'CR_Registrations').CR_Registrations)

for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
ax.invert_yaxis()
 
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5, 
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
ax.set_title('Конверсия в регистрацию по гео',
             loc ='left');
```

* Конверсия в регистрацию хуже всего была у кампании с гео "cis"
